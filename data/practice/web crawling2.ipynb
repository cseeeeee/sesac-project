```python
cleaned_top_movies=[]
for movie in top_movies:
  code=movie['code']
  title=movie['title']
  year=movie['year']
  ratings_avg=movie['ratings_avg']
  nations=movie['nations']
  genres=movie['genres']
  movie_dict={
    'code':code,
    'title':title,
    'year':year,
    'ratings_avg':ratings_avg,
    'nations':nations,
    'genres':genres
  }
  cleaned_top_movies.append(movie_dict)
```

```python
cleaned_top_movies
```

#### 문제3: `cleaned_top_movies`를 데이터 프레임으로 만들고, `watcha_top_movies.csv` 파일로 저장하세요.

```python
top_movies_df = pd.DataFrame(cleaned_top_movies)

top_movies_df=top_movies_df.drop_duplicates(['code'])
top_movies_df.reset_index().iloc[:, 1:]
```

```python
top_movies_df.to_csv('watcha_top_movie.csv',index=False)
```
